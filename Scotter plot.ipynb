```python
df_google_play_store_apps= pd.read_csv(r"E:\jupyter\archive\googleplaystore.csv" ,nrows = 1000)#nrows is uesd to get only 1000 recors
df_google_play_store_apps.shape
```

```python
df_google_play_store_apps
```

```python
x = df_google_play_store_apps["Rating"]   
y = df_google_play_store_apps["Reviews"]

plt.scatter(x,y)
plt.title("google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
x = df_google_play_store_apps["Rating"]   
y = df_google_play_store_apps["Reviews"]

plt.figure(figsize = ( 16,9))   # use to large the size of scatter plot
plt.scatter(x,y ,c = "r" ,marker ="o",s = 100, alpha = 0.5 ,edgecolor ="k")  #alpha use to change trasferency of dot
plt.title("google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
x = df_google_play_store_apps["Rating"]   
y = df_google_play_store_apps["Reviews"]

plt.figure(figsize = ( 16,9))   # use to large the size of scatter plot
plt.scatter(x,y ,c = "r" ,marker ="o",s = 100, alpha = 0.5 ,edgecolor ="k") 

plt.scatter(x,df_google_play_store_apps["Installs"],c = "b" ,marker ="*",s = 100, alpha = 0.4 ,edgecolor ="y") 

plt.title("google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```
