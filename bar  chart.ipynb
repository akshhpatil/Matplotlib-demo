# ploting bar chat 


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#Data set of indian artificial intelligence production class

classes = ["python","AI","ML","R","DS"]

cla1_stud = [ 30,20,10,20,30]
cla2_stud = [ 15,30,15,20,10]
cla3_stud = [ 10,20,16,34,23]
```

```python
plt.bar(classes,cla1_stud )
```

```python
plt.barh(classes,cla1_stud)
```

```python
plt.bar(classes,cla1_stud, 
        width = 0.2,
        color = "y", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 1 stud",visible = True ,
        animated =True,
        capstyle ='round'

       )
plt.figure(figsize = (15,15)) 
```

```python
style.use("ggplot")

plt.figure(figsize = (15,9))

plt.bar(classes,cla1_stud, 
        width = 0.5,
        color = "y", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 1 stud",visible = True ,
        animated =True,
        capstyle ='round',
        linewidth = 5

       )

plt.title("Bar chat of AIP class" ,fontsize = 18 ,color = "b")
plt.xlabel("classes", fontsize = 15,color = "b") 
plt.ylabel("no of student" ,fontsize = 15, color = "b")
 
```

```python
plt.figure(figsize = (15,9))



plt.bar(classes,cla1_stud, 
        width = 0.3,
        color = "y", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 1 stud",visible = True ,
       

       )

plt.bar(classes,cla2_stud, 
        width = 0.3,
        color = "m", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 2 stud",
        visible = True ,
       

       )

plt.title("Bar chat of AIP class" ,fontsize = 18 ,color = "b")
plt.xlabel("classes", fontsize = 15,color = "b") 
plt.ylabel("no of student" ,fontsize = 15, color = "b")
```

```python
plt.figure(figsize = (15,9))

classes_index = np.arange(len(classes))

width = 0.2



plt.bar(classes_index,cla1_stud, 
        width,
        color = "y", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 1 stud"
       

       )

plt.bar(classes_index + width ,cla2_stud, 
         width,
        color = "m", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 2 stud"
     
       

       )
plt.bar(classes_index + width + width ,cla3_stud, 
        width,
        color = "b", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 3 stud"
     
        )


plt.xticks( classes_index + width ,classes,rotation = 20)

plt.title("Bar chat of AIP class" ,fontsize = 18 ,color = "b")
plt.xlabel("classes", fontsize = 15,color = "b") 
plt.ylabel("no of student" ,fontsize = 15, color = "b")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,9))

classes_index = np.arange(len(classes))

width = 0.2



plt.barh(classes_index,cla1_stud, 
        width,
        color = "y", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 1 stud"
       

       )

plt.barh(classes_index + width ,cla2_stud, 
         width,
        color = "m", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 2 stud"
     
       

       )
plt.barh(classes_index + width + width ,cla3_stud, 
        width,
        color = "b", 
        edgecolor = "k",
        linestyle = "--",
        label = "class 3 stud"
     
        )


plt.yticks( classes_index + width ,classes,rotation = 20)

plt.title("Bar chat of AIP class" ,fontsize = 18 ,color = "b")
plt.ylabel("classes", fontsize = 15,color = "b") 
plt.xlabel("no of student" ,fontsize = 15, color = "b")
plt.legend()
plt.show()
```
