# ploting histograms

```python
import matplotlib.pyplot as plt 
import numpy as np 
import random 
```

```python
m_stud_age  = np.random.randint( 18,45, (100))  #   age between  18 to 45 and 100 student 
data_sci_stude  =np.random.randint( 20,40, (100))   #   age between  20 to 40 and 100 student 
 
    
```

```python
print(m_stud_age)
print(data_sci_stude)
```

```python
plt.hist(m_stud_age )

plt.title("Machine learning student histogram")
plt.xlabel("Student age ")
plt.ylabel("no of student age")
plt.show()
```

```python
bins = [ 15,20,25,30,35,40,45]
```

```python
plt.hist(m_stud_age,bins ,rwidth = 0.8 , color = 'm' ,label = "ML student")

plt.title("Machine learning student histogram")
plt.xlabel("Student age ")
plt.ylabel("no of student age")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,9))   # increase the value of histogram

plt.hist([m_stud_age,data_sci_stude], bins ,rwidth = 0.7 , color = ['m','y'] ,label = ["ML student","data sc stud "])

#plt.hist(data_sci_stude,bins ,rwidth = 0.7 , color = 'y' ,label = "data sc stud")


plt.title("Machine learning student histogram")
plt.xlabel("Student age ")
plt.ylabel("no of student age")
plt.legend()
plt.show()
```
