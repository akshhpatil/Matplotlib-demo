# pie chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["python","R","ML","AI","DS"]

class1_stud = [ 45,15,35,25,30]

plt.pie([1])
plt.show()
```

```python
classes = ["python","R","ML","AI","DS"]

class1_stud = [ 45,15,35,25,30]

plt.pie(class1_stud , labels = classes )

plt.show()
```

```python
classes = ["python","R","ML","AI","DS"]

class1_stud = [ 45,15,35,25,30]

explode = [0.03,0,0.1,0,0]

plt.pie(class1_stud , labels = classes ,explode = explode )

plt.show()
```

```python
classes = ["python","R","ML","AI","DS"]

class1_stud = [ 45,15,35,25,30]

explode = [0.03,0,0.1,0,0]

colors =["c", "b","r","y","g" ]

textprops = {"fontsize" : 15}

plt.pie(class1_stud , 
        labels = classes ,
        explode = explode ,        #silce of pie chart 
        colors = colors ,
        autopct = "%0.2f%%",       # when you want to show % in pie chart
        shadow = True, 
        radius = 1.4 ,              # increse in size
        startangle = 90,  
        textprops = textprops       #increae the text size
        
       )

plt.show()
```

```python
wedgeprops = { "linewidth": 3, "width":1, "edgecolor": "k"}

plt.pie(
        class1_stud , 
        labels = classes ,
        explode = explode ,        
        colors = colors ,
        autopct = "%0.2f%%",      
        shadow = True, 
        radius = 1.1 ,              
        startangle = 90,  
        textprops = textprops ,
        pctdistance=0.6,
        labeldistance=1.4,
        counterclock=True,
        wedgeprops=wedgeprops,
        center=(2, 3),
        frame=False,
        rotatelabels=False
   
        
       )

plt.legend(loc= 4)

plt.show()
```
