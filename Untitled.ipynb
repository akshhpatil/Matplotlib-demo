# Image show in matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread("pie_chart.png")
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
single_channel = img [:,:,2]
plt.axis("off")
plt.imshow(single_channel,cmap ="hot")
plt.colorbar()
plt.show()
```
